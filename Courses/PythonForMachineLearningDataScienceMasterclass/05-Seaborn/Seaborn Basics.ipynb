# Seaborn Basics

- Seaborn is built off of matplotlib but gives us convenience.
- We can use matplotlib methods to modify what we are looking at.

## Scatter Plots
- Scatter plots show the relationship between two continuous features.
- Continuous features are numeric variables that can take any numbr of values betweeen any two values.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
sns.scatterplot(
    x = 'salary',
    y = 'sales',
    data = df
);
```

```python
# What if we wanted to edit figure size? we can use matplotlib.
plt.figure(figsize=(12, 4), dpi=200)

# Scatterplot func.
sns.scatterplot(
    # X feature.
    x = 'salary',
    # Y feature.
    y = 'sales',
    # Dataframe.
    data = df,
    # Add a hue to the plot (colour the plots by this category)
    hue='level of education',
    # hue='salary' numerical would show a gradient.
    palette = 'Dark2' # change the colour map (look up matplotlib colour maps)
);
```

```python
# What if we wanted to edit figure size? we can use matplotlib.
plt.figure(figsize=(12, 4), dpi=200)

# Scatterplot func.
sns.scatterplot(
    # X feature.
    x = 'salary',
    # Y feature.
    y = 'sales',
    # Dataframe.
    data = df,
    # Change the size of plots as salary increases
    size='salary'
    # s = 200 change general size of all points.
    # alpha = 0.5 add a fade to the graph plots.
);
```

```python
# What if we wanted to edit figure size? we can use matplotlib.
plt.figure(figsize=(12, 4), dpi=200)

# Scatterplot func.
sns.scatterplot(
    # X feature.
    x = 'salary',
    # Y feature.
    y = 'sales',
    # Dataframe.
    data = df,
    # Automatically shape plots based of the style of the record.
    style= 'level of education',
    # Also show the different colour for each shape.
    hue = 'level of education'
);

# Save fig using matplotlib.
plt.savefig('newguy.jpg')
```

## Distribution Plots

- Display single continuos feature that help show deviation.
- Rug plot is the most simple (adds a tick for each datapoint), y axis is not interperetable.
- Histogram shows how many ticks there are for a range. shows how many values are in each "bin"
- y axis for histogram could be a percentage of all points.
- We can add a KDE line on the historgram (Kernel Density Estimation) a method of estimating a continuos probability curve for a finite data sample. (most common is the gausien (normal distribution))
