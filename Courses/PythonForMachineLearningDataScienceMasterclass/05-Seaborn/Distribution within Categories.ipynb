```python
# Main plots for categories (boxplot, violin plot, swarmplot)
# boxplot most important
# it shows statistical paramaters based on quartiles
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
# Pyplot figure mod.
plt.figure(figsize=(10, 4), dpi=200)

# BOXPLOT
sns.boxplot(
    data=df, 
    y='math score', 
    # Split by category!
    x='test preparation course'
)
```

```python
# Pyplot figure mod.
plt.figure(figsize=(10, 4), dpi=200)

# BOXPLOT (mapping a continuos feature against a categorical one)
sns.boxplot(
    data=df,
    # Could swap x and y to show landscape!
    y='reading score', 
    # Split by category!
    x='parental level of education',
    # Apply a hue to show a third dimension of info
    hue='test preparation course',
    palette='Set2'
)

plt.legend(bbox_to_anchor=(1.02, 1))
```

```python
# Pyplot figure mod.
plt.figure(figsize=(10, 4), dpi=200)

# BOXPLOT (mapping a continuos feature against a categorical one)
sns.boxplot(
    data=df,
    # Could swap x and y to show landscape!
    x='reading score', 
    # Split by category!
    y='parental level of education',
    # Apply a hue to show a third dimension of info
    hue='test preparation course',
    palette='Set2'
)

plt.legend(bbox_to_anchor=(1.02, 1))
```

```python
# Pyplot figure mod.
plt.figure(figsize=(10, 4), dpi=200)

# VIOLINPLOT with hue
sns.violinplot(
    data=df,
    # Could swap x and y to show landscape!
    y='reading score', 
    # Split by category!
    x='parental level of education',
    # Apply a hue to show a third dimension of info
    hue='test preparation course',
    palette='Set2'
)

plt.legend(bbox_to_anchor=(1.02, 1))
```

```python
# Pyplot figure mod.
plt.figure(figsize=(10, 4), dpi=200)

# VIOLINPLOT with split
sns.violinplot(
    data=df,
    # Could swap x and y to show landscape!
    y='reading score', 
    # Split by category!
    x='parental level of education',
    # Apply a hue to show a third dimension of info
    hue='test preparation course',
    # Show subcategories next to eachother (1 half is 1 half and 1 half is the other)
    split='true',
    palette='Set2'
)

plt.legend(bbox_to_anchor=(1.02, 1))
```

```python
# Pyplot figure mod.
plt.figure(figsize=(10, 4), dpi=200)

# VIOLINPLOT with quartile
sns.violinplot(
    data=df,
    # Could swap x and y to show landscape!
    y='reading score', 
    # Split by category!
    x='parental level of education',
    # Apply a hue to show a third dimension of info
    hue='test preparation course',
    # Show subcategories next to eachother (1 half is 1 half and 1 half is the other)
    split='true',
    inner='quartile',
    palette='Set2'
)

plt.legend(bbox_to_anchor=(1.02, 1))
```

```python
# Pyplot figure mod.
plt.figure(figsize=(10, 4), dpi=200)

# VIOLINPLOT with stick
sns.violinplot(
    data=df,
    # Could swap x and y to show landscape!
    y='reading score', 
    # Split by category!
    x='parental level of education',
    # Apply a hue to show a third dimension of info
    hue='test preparation course',
    # Show subcategories next to eachother (1 half is 1 half and 1 half is the other)
    split='true',
    inner='stick',
    palette='Set2'
)

plt.legend(bbox_to_anchor=(1.02, 1))
```

```python
# Pyplot figure mod.
plt.figure(figsize=(8, 4), dpi=200)

# SWARMPLOT
sns.swarmplot(
    data=df,
    x='math score',
    y='gender',
    size=2, # modify size of dots.,
    hue='test preparation course',
    dodge=True # Split up the hue diagrams
)

plt.legend(bbox_to_anchor=(1.35, 1.05))
```

```python
# BOXENPLOT
sns.boxenplot(
    x='math score',
    y='test preparation course',
    data=df,
    hue='gender'
)
```
