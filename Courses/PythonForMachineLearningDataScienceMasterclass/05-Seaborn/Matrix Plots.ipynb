## Matrix Plots
- Visual equivalent of displaying a pivot table

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('country_table.csv')
df = df.set_index('Countries') # Need a labeled index
df
```

```python
# HEATMAPS
# - Each value in heatmap should be of the same unit.
# df = df.drop('Life expectancy', axis=1) # Life expetency was not a rate.

sns.heatmap(
    data=df,
    linewidth=2,
    annot=True, # annotate with values
    cmap='viridis'
)
```

```python
# CLUSTERMAP
sns.clustermap(
    data=df,
    linewidth=2,
    annot=True, # annotate with values
    cmap='viridis'
)
```
