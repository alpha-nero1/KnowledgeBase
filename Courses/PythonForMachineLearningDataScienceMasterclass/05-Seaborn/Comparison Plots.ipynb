```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 2-D versions of the plots we have looked at so far.
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
# JOINTPLOT
# - adds in histograms as part of scatterplot.

sns.jointplot(
    data=df,
    x='math score',
    y='reading score',
    alpha=0.2 # Show distribution in scatter
)
```

```python
sns.jointplot(
    data=df,
    x='math score',
    y='reading score',
    # Can change the type of join plot!
    kind='hex'
)
```

```python
sns.jointplot(
    data=df,
    x='math score',
    y='reading score',
    kind='hist' # HISTOGRAM INNER
)
```

```python
sns.jointplot(
    data=df,
    x='math score',
    y='reading score',
    kind='kde', # KDE INNER
    shade=True
)
```

```python
sns.jointplot(
    data=df,
    x='math score',
    y='reading score',
    # Jointplot with hue!
    hue='gender'
)
```

```python
# PAIRPLOT
# - warning can be CPU and RAM intensive for each column
sns.pairplot(
    data=df, # Grabbed all numerical columns and showed distributions + relationships of each.
    hue='gender', # Can add a hue
    corner=True # Remove duplicates
)
```
