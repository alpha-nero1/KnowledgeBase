# Why Machine Learning

- In general is the study of statistical computer algorithms that improve automatically through data.
- is a subset for AI
- What kind of problems can it solve: Credit Scoring, Insurance Risk, Price Forecasting, Spam Filtering, Customer Segmentation, Entity Classification

## Structure of ML problem framing

- Given features from a data set obtain a desired label.
- ML algorithms are often called "estimators" as they estimate the desired label or output.

## So why do we use it?

- Some problems are only possible using ML.
- major caveat is that we NEED good existing data.
- majority of time is spent on preparaing and massagind the data into the form you want.

# Different Types of ML

Two main types are:

## Supervised Learning

- Uses historical and labeled data, the ML model PREDICTS a value
- Known results and data from the past.
- labeled data means that we KNOW the features of the data and past prediciton labels.

### Two main label types

- Categorical value to predict (CLASSIFICATION task) (E.g predict the category, handwriting detection)
- Continuos value to predict (REGRESSION task) (E.g. predict the price, electricity loads, test scores)

## Unsupervised Learning

- Applied to unlabeled data, the ML model discovers possible patterns in the data.
- Desired output is unknown.
- Trying to group and label data without a label.
- Downside is that there is no historical correct label, it's hard to evaluate performance.

# Supervised ML

PREDICTING AN OUTCOME

scikit-learn used for ML models.

- Starts with collecting and organising a data set on history.
- Create a data product that can help us with predicting.
- x: features, y: label
- label is a term for what we are trying to predict
- features are known components in the data we are predicting the label from.

- Once we have identified features and label, we split the data into a training set and a test set.
- We split by train and test so that we can verify that our model is high performant.

## Training Set

- The data we use to train the model.


## Testing Set

- The data we use to check against the trained model to test that predicted values are accurate.
- We provide only the features not the labels when testing the algorithm.
- The algorithm could not have memorized data it has never seen which is why we do this, we can see that predictions are accurate or not.

- Common split is 70% training, 30% test.


What happend if performance is not great?
- We can adjust the models paramaters and try again.

## Beggining to End, the Process

1. We get the x and y data (features + label)
2. split the data into test and training set.
3. Fit ML model on training data set.
4. Evaluate model performance given test set.
5. adjust model params if performance is poor.
6. once satisified, deploy model to the real world.
