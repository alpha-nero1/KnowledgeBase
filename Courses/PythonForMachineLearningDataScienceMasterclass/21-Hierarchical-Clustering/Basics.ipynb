```python
import numpy as np
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../DATA/cluster_mpg.csv')
```

```python
df
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
scaler = MinMaxScaler() # Better for heirarchical clustering (range of 0-1)
```

```python
df['origin'].value_counts()
```

```python
df_w_dummies = pd.get_dummies(df.drop('name', axis=1))
```

```python
df_w_dummies
```

```python
scaled_data = scaler.fit_transform(df_w_dummies)
```

```python
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=df_w_dummies.columns) # Scaled data is back as a DataFrame!
```

```python
scaled_df
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(scaled_df, cmap='magma')
```

```python
len(df)
```

```python
sns.clustermap(scaled_df, row_cluster=False) # CPU intensive, we can see feature similarities.
```

```python
# OR..
scaled_df.corr()
```

```python
sns.clustermap(scaled_df, col_cluster=False) # Row clustering...
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=4)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize=(10, 6), dpi=200)
sns.scatterplot(data=df, x='mpg', y='weight', hue=cluster_labels, palette='viridis') # Can pass in cluster labels as a hue!!
```

```python
max_distance_between_two_points = np.sqrt(len(scaled_df.columns)) # Max distance between two points because of the MinMaxScaler
max_distance_between_two_points
```

```python
model = AgglomerativeClustering(
    n_clusters=None, # Now we need to define distance threshold
    distance_threshold=0
)
```

```python
cls = model.fit_predict(scaled_df)
```

```python
cls # Becuase threshold was 0 there will be a cluster for each individual point.
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix # The linkage is returning two clusters and distance between them (then the number of points in the cluster)
# row = c1, c2, distance, total_points
# math info we need for dendrogram
```

```python
plt.figure(figsize=(20, 10), dpi=200)
# Useful to see hierarchy cutoff
dendro = dendrogram(linkage_matrix)
```

```python
plt.figure(figsize=(20, 10), dpi=200)
dendro = dendrogram(linkage_matrix, truncate_mode='lastp', p=10)
```

```python
plt.figure(figsize=(20, 10), dpi=200)
dendro = dendrogram(linkage_matrix, truncate_mode='level', p=3)
```
