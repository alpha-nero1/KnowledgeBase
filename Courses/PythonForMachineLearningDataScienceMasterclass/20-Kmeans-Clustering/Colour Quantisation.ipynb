# Colour Quantisation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.image as mpimg # lib for working with images
```

```python
image_arr = mpimg.imread('../DATA/palm_trees.jpg') # Read in an image
```

```python
image_arr # Array of RGBs
```

```python
image_arr.shape # H, W, C
```

```python
plt.figure(dpi=200)
plt.imshow(image_arr) # Programatically show the image from the array
```

```python
(h, w, c) = image_arr.shape
```

```python
h
```

```python
w
```

```python
c
```

```python
image_as_arr2d = image_arr.reshape(h * w, c)
```

```python
image_as_arr2d.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=6)
```

```python
clusters = model.fit_predict(image_as_arr2d)
```

```python
clusters
```

```python
model.cluster_centers_ # Cluster centers
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
quantised_img = np.reshape(rgb_codes[clusters], (h, w, c))
```

```python
quantised_img
```

```python
plt.figure(dpi=200)
plt.imshow(quantised_img) # Programatically show the image from the array
```
