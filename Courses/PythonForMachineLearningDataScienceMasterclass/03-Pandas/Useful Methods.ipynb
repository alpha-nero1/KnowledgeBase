## apply() function

```python
# The most useful.
# Allows us to apply any custom python function to every row in a series

import pandas as pd
import numpy as np
df = pd.read_csv('tips.csv')
df.head()
```

```python
# Get the last 4 digits of CC number.
df.info()
```

```python
str(12345675624)[-4:]
```

```python
def last_four(number):
    return str(number)[-4:]
```

```python
last_four(6758585)
```

```python
# Apply func to every row

df['CC Number'].apply(last_four) # So cool!!!
```

```python
# Add new column with the derived data
df['Last Four CC Number'] = df['CC Number'].apply(last_four)
```

```python
df
```

```python
df['total_bill'].mean() # Get the mean proce for a bill
```

```python
# look: numerical to nominal function
def yelp(price):
    if price < 10: return '$'
    if price < 30: return '$$'
    return '$$$'
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```

```python
df
```

## .apply() with multiple columns

```python
lam = lambda num: num * 2
```

```python
df['total_bill'].apply(lambda bill: bill * 2)
```

```python
# Use apply to receive two columns.
#
df.head()
```

```python
# Name same as columns.
def quality_of_tip(total_bill, tip):
    if tip / total_bill > 0.25: return "Generous"
    return "Other"
```

```python
# We can use a lambda on apply to receive the WHOLE dataframe row and do what we want with it.
# We specify axis=1 for columns.
df['tip_quality'] = df[['total_bill', 'tip']].apply(lambda df: quality_of_tip(df['total_bill'], df['tip']), axis = 1)
```

```python
df[:100]
```

```python
# We can make these functions run a lot faster using vectorise
# vectorize purpose is to transform functions that are NOT numpy aware (has no idea that we are broadcasting returns)
df['tip_quality'] = np.vectorize(quality_of_tip)(df['total_bill'], df['tip'])
```

```python
df
```

## Timeit Code Efficiency Measuring

```python
# We can test this using the timeit module!!
import timeit
```

```python
# Snippet run only once for statements
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('tips.csv')
def quality_of_tip(total_bill, tip):
    if tip / total_bill > 0.25: return "Generous"
    return "Other"
'''
```

```python
# Snippets we want to be measured
st_one = '''
df['tip_quality'] = df[['total_bill', 'tip']].apply(lambda df: quality_of_tip(df['total_bill'], df['tip']), axis = 1)
'''

st_two = '''
df['tip_quality'] = np.vectorize(quality_of_tip)(df['total_bill'], df['tip'])
'''

# Runs for multiple loops and returns average
```

```python
# Measure number one
timeit.timeit(setup=setup, stmt=st_one, number=1000)
```

```python
# Measure number two, Holy whattttt!!! vectorize is really quick!
timeit.timeit(setup=setup, stmt=st_two, number=1000)
```
