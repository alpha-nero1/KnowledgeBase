# Styling Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Legends

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
x = np.linspace(0, 11, 10)

# Two plots on one axis.
# With ach plot call we need to pass a label.
ax.plot(x, x, label='X vs X')
ax.plot(x, x** 2, label='X vx X^2')

# Will include the legend.
# Can pass in loc arg to change legend location
# ax.legend(loc='lower left')
# OR manually place the loaction using a tuple.
ax.legend(loc=(1.02, 0.88))
```

## Line styling

```python
x = np.linspace(0, 11, 10)

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color='blue', label='X') # Can pass in colour!
ax.plot(x, x**2, color='#650dbd', label='X ^ 2') # Can pass in hex colour!
ax.plot(x, x**3, color='#e6c017', label='x ^ 3') # Can pass in hex colour!
ax.plot(x, x**4, color='orange', label='x ^ 4', lw=10) # Can change line width
ax.plot(x, x**4.5, color='red', ls='-.') # Can change line styles! `--`, `-`, `:`, `-.`

# What if I want to actually plot the points?
ax.plot(x, x**4.6, color='brown', ls='--', marker='o') # Marker string codes: matplotlib.markers
ax.plot(x, x**4.7, color='yellow', ls='--', marker='+', ms=20) # ms (marker size) to change how large the markers are.

# Full customisation:
ax.plot(
    x, 
    x**4.8, 
    color='purple', 
    linewidth=2, 
    marker='o', 
    ls='--', 
    markersize=10,
    markerfacecolor='red',
    markeredgewidth=5,
    markeredgecolor='orange'
)

ax.legend()
```
