```python
# Lets see them in action!
fg = plt.figure() # Create a blank canvas

# Add set of axes
axes_one = fg.add_axes([0, 0, 1, 1])

# We can plot two things on the one figure!!!!
axes_one.plot(x, y)
axes_one.plot(a, b)
```

```python
# Can add as many figure axes as you want.
# Lets see them in action!
fg = plt.figure() # Create a blank canvas

# Large axes
axes_one = fg.add_axes([0, 0, 1, 1])

# Small axes
axes_two = fg.add_axes([0.2, 0.2, 0.5, 0.5]) # Insert axes.

# We can plot two things on the one figure!!!!
axes_one.plot(x, y)
axes_one.plot(a, b)
```

```python
# Can add as many figure axes as you want.
# Lets see them in action!
fg = plt.figure() # Create a blank canvas

# Large axes
axes_one = fg.add_axes([0, 0, 1, 1])

# Small axes, gets added INSIDE the larger one.
axes_two = fg.add_axes([0.2, 0.5, 0.25, 0.25]) # Insert axes.

# We can plot two things on the one figure!!!!
axes_one.plot(x, y)
axes_one.plot(a, b)

# Plot on the smaller graph
axes_two.plot(a, c)
```

```python
# Can add as many figure axes as you want.
# Lets see them in action!
fg = plt.figure() # Create a blank canvas

# Large axes
axes_one = fg.add_axes([0, 0, 1, 1])

# Add labels.
axes_one.set_xlim(0, 8)
axes_one.set_ylim(0, 8000)
axes_one.set_ylabel('A')
axes_one.set_xlabel('B')
axes_one.set_title('Power of four')

# Small axes, gets added INSIDE the larger one.
axes_two = fg.add_axes([0.2, 0.5, 0.25, 0.25]) # Insert axes.

# Zoomed in!
axes_two.set_xlim(1, 4)
axes_two.set_ylim(0, 20)
axes_two.set_ylabel('A')
axes_two.set_xlabel('B')
axes_two.set_title('Zoomed in')

# We can plot two things on the one figure!!!!
axes_one.plot(a, b)
axes_one.plot(a, b)

# Plot on the smaller graph
axes_two.plot(a, c)
```

```python
# Figure size and DPI
fg = plt.figure(dpi=200, figsize=(10, 3))

axes_one = fg.add_axes([0, 0, 1, 1])
axes_one.plot(a, b)

fg.savefig('newfigure.png')
```

```python
# Figure size and DPI
fg = plt.figure(dpi=200, figsize=(2, 2))

axes_one = fg.add_axes([0, 0, 1, 1])
axes_one.plot(a, b)

## bb = bounding box
fg.savefig('newfigure.png', bbox_inches='tight') # bbox_inches='tight' will include x and y axes tick numbers.
```
