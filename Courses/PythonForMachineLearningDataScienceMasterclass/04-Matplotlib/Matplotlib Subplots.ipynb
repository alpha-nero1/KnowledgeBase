```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0, 10, 11)
```

```python
b = a ** 3
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
# Calling subplots allows us to provide cols and rows to return a nparray of axes. This lets us bootsrap a bunch of plots.
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# A numpy array.
axes
```

```python
# Load the data!
axes[0].plot(x, y)
axes[1].plot(a, b) 
```

```python
fig
```

```python
# Could do just a stack like:
fig, axes = plt.subplots(nrows=3, ncols=1)

axes[0].plot(x, y)
axes[1].plot(a, b)

# Ensure no tickers overlap
fig.tight_layout()
```

```python
## CUSTOM SPACING
# Could do just a stack like:
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(4, 10), dpi=200) # Has same fig params

axes[0][0].plot(x, y)
axes[0][1].plot(a, b)
axes[1][0].plot(x, y)
axes[1][1].plot(a, b)

# Add labels to the individual axes
axes[1, 1].set_ylabel('Hello')
axes[1, 1].set_xlabel('Hey')

# Figure level title.
fig.suptitle('Figure level')

# Custom spacing here.
fig.subplots_adjust(wspace=0.6, hspace=0.6); # Adjust spaacing... expressed as a fraction of the average axes hight

fig.savefig('newsubplots.png', bbox_inches='tight')
```
